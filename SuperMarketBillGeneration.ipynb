```python
import math
from datetime import datetime

name = input('Enter your name: ')

def bill_cal():
    a = True
    total_price = 0
    list_item_price = []
    d = {
        "apples": 4.00,
        "bananas": 1.50,
        "milk": 3.50,
        "bread": 3.25,
        "eggs": 4.00,
        "cheese": 7.00,
        "tomatoes": 4.50,
        "potatoes": 3.00,
        "chicken": 10.00,
        "rice": 6.00,
        "carrots": 2.50,
        "onions": 2.00,
        "beef": 12.00,
        "pasta": 3.00,
        "olive oil": 12.00,
        "butter": 6.00,
        "yogurt": 4.00,
        "oranges": 5.00,
        "lettuce": 2.50,
        "cucumbers": 2.00,
        "spinach": 3.50,
        "grapes": 5.00,
        "strawberries": 6.50,
        "blueberries": 5.50,
        "peanut butter": 5.00,
        "jam": 4.00,
        "flour": 3.50,
        "sugar": 3.00,
        "coffee": 10.00,
        "tea": 6.00,
        "salt": 1.50,
        "pepper": 3.50,
        "garlic": 2.00,
        "ginger": 3.00,
        "broccoli": 4.00,
        "cauliflower": 4.50,
        "zucchini": 3.00,
        "bell peppers": 3.50,
        "avocados": 2.50,
        "mushrooms": 5.00,
        "honey": 10.00,
        "cereal": 6.00
    }

    while a:
        print('''
            1. List of items
            2. Choose item
            3. Exit
        ''')

        choice = input('Enter your choice: ')

        if choice == '1':
            c = 1
            for i, j in d.items():
                print(f'\t{c}. {i} ‚Çπ{j}')
                c += 1

        elif choice == '2':
            q = True
            while q:
                print("Press 'q' to return to the main menu.")
                item = input('Enter item: ').lower()
                if item in d:
                    qnt = input('Enter Quantity: ')
                    if qnt.isdigit():
                        qnt = int(qnt)
                        price = float(d[item] * qnt)
                        list_item_price.append((item, qnt, price))
                        total_price += price
                    else:
                        print('Invalid Quantity Input.')
                elif item == 'q':
                    q = False
                else:
                    print('Item not present.')

        elif choice == '3':
            a = False
        else:
            print('Invalid Input. Please Enter a valid choice.')

    return total_price, 'Thank you, please visit again.', list_item_price

total, msg, item_price = bill_cal()
if total != 0:
    print('\n')
    print('''
                JP Stores
            BTM, Bangalore, 560076

    ''')
    print('Customer Name:', name, '\t', 'Date:', datetime.now())
    print(20 * '==')

    for j in item_price:
        print(f'Item: {j[0]}\tQuantity: {j[1]}\tPrice: ‚Çπ{j[2]:.2f}')

    gst = total * 0.1
    gst = math.ceil(gst)
    print(18 * '==')
    print(f'GST: ‚Çπ{gst:.2f}')
    print(f'Total payable amount: ‚Çπ{(gst + total):.2f}')
    print(18 * '==')
    print('üôèüôèüôè ', msg, ' üôèüôèüôè')
else:
    print("Hey, you didn't buy anything... Please buy something you need.")
```
